Exercise 1 : Calculating The Mean and the median

```python
import pandas as pd

data = pd.Series([2,4,6,8,10])
# print("Original Data :", data)

mean = data.mean()
median = data.median()
# print("Mean of the data:", int(mean))
print("Median of the data:", int(median))
```

#Calculating the mode

```python
data = pd.Series([2,2,2,3,3,3,4,6,8,10])

# mode = data.mode()
mode = data.mode()[0]
print("Mode of the data", mode)
```

#Exercise : Calculating Range

```python
data = pd.Series([2,4,5,8,10])

range = data.max() - data.min()
print("Range of the data", range)
```

##Exercise : Calculating Variance and Standard Deviation

```python
data = pd.Series([2,4,5,8,10])

standard_deviation = data.std()
variance = data.var()
print("Standard Deviation of the data", standard_deviation)
print("Variance of the data", variance)
```
